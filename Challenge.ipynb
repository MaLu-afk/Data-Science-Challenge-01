### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



Vamos a visualizar el ingreso total de cada tienda.

```python
# Importamos matplotlib para graficar y sea más comprensible
import matplotlib.pyplot as plt
```

```python
ingreso_total = {
    'ingreso_total_tienda1' : tienda['Precio'].sum(),
    'ingreso_total_tienda2' : tienda2['Precio'].sum(),
    'ingreso_total_tienda3' : tienda3['Precio'].sum(),
    'ingreso_total_tienda4' : tienda4['Precio'].sum()
}
```

```python
# Verificamos el ingreso total de cada tienda.
ingreso_total
```

```python
# Extraemos claves y valores
x = list(ingreso_total.keys())
y = list(ingreso_total.values())

plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue', edgecolor='black')
plt.title('Ingresos por tiendas')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# Ahora, vamos a visualizar qué tanto se dispersan el número de artículos vendidos por categoría y tienda:
diccionarios = [num_productos_vendidos1, num_productos_vendidos2, num_productos_vendidos3, num_productos_vendidos4]
nombres_series = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]

plt.figure(figsize=(12, 6))
colores = ['blue', 'red', 'green', 'purple']

for i, dicc in enumerate(diccionarios):
    categorias = list(dicc.keys())
    valores = list(dicc.values())
    plt.scatter(x=categorias, y=valores, color=colores[i], label=nombres_series[i], s=100, alpha=0.7)

plt.title("Comparación de Ventas por Categoría", fontsize=14)
plt.xlabel("Categorías", fontsize=12)
plt.ylabel("Número de Productos Vendidos", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()
```

Si nos detenemos a pensar por un momento, puesto que la tienda 4 presenta menor ingreso, podemos sumir que es porque tiene menos venta de productos. Pero en la gráfica se muestra que el número de productos vendidos se mantiene a comparación de las otras tiendas por categoría a excepción de la categoría "Electrodomésticos" que hay un número de ventas menor en la tienda 4.
